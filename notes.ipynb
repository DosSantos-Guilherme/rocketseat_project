# HTML
It is a marking language, structured by tags that define the content of a web page.
# CSS
It defines the style of a element, the definitions works in a cascade, where the last definition overwrittes the previous one, respecting its order and specificity.
## Border-box
By default the size of an element is defined by the sum of padding, margin, and content. That is not intuitive and developers usually set for the whole css the border-box type of sizing. Where the actual element size is defined only by content. That is done by:
```
*{
  box-sizing:border-box;
}
``` 
# Box model
It is a way of representing a HTML element as a rectilinear box with marging, padding, and content.
# Block vs inline elements
By default some structures are already defined as block or as inline elements. 
## Block
The block elements ocopy the whole line, not letting anything to be at their side.
## Inline
The inline elements only ocupy it's content which allows some other inline structures to be at their side.
### Issue generated by the default block or inline structures
That creates a behavior where some commands might not work. For instance, if you use margin: auto for a blocked element it will get centralized, while it would not work for a inline one. 
### Solution
If the upward container is inline you can send styles aiming for aligning whatever is inside. So you can align by setting `text-align:center;`

# Flex
It allows elements to fit in one direction, occupying all there is to occupy within this content. The default is horizontal, but you can change it.
The display flex creates a line in the middle of the content, named main. You can use that to align vertically.

# position: relative
That will not alter the element itself, but all absolute elements within it will have absolute position respective to the upward element.
